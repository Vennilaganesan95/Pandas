```python
import pandas as pd

# Sample dataset
data = {
    "Student": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Course": ["Computer Science", "Physics", "Mathematics", "Computer Science", "Physics", "Mathematics"],
    "Age": [22, 22, 26, 23, 20, 24],
    "Grade": [65, 88, 98, 55, 65, 80],
    "City": ["Chicago", "New York", "Chicago", "New York", "Boston", "Chicago"]
}

df = pd.DataFrame(data)

print("Student Data:\n", df)

df_selected = df[["Student", "Grade"]] 
print(df_selected)

high_grade_students = df[df["Grade"] > 80] 
print("\n",high_grade_students)

cs_students = df[df["Course"].isin(["CS"])] 
print("\n",cs_students)

age_filtered_students = df[df["Age"].between(22, 26)]
print("\n",age_filtered_students)

df_indexed = df.set_index("Student") 
print("\n",df_indexed)

df_reset = df_indexed.reset_index() 
print("\n",df_reset)
```
