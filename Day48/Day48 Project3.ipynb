```python
import pandas as pd

data = {
    "Employee_ID": [101, 102, 103, 104, 105, 106, 107, 108],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen"],
    "Department": ["IT", "Finance", "HR", "IT", "Finance", "Marketing", "IT", "HR"],
    "Salary": [6000, 7500, 4500, 5200, 8000, 4000, 9000, 4800],
    "Experience": [3, 8, 12, 6, 10, 4, 9, 7],
    "City": ["New York", "Chicago", "Los Angeles", "New York", "Houston", "Chicago", "New York", "Chicago"],
    "Age": [25, 35, 40, 28, 32, 31, 29, 45]  # Added for Task 13
}

df = pd.DataFrame(data)

# 1. Load the dataset into Pandas DataFrame
print("Employee Dataset:\n", df, "\n")

# 2. Display first five and last five rows
print("First 5 rows:\n", df.head(), "\n")
print("Last 5 rows:\n", df.tail(), "\n")

# 3. Select and display only Name and Department
print("Name and Department:\n", df[["Name", "Department"]], "\n")

# 4. Employees in IT Department
print("Employees in IT:\n", df[df["Department"] == "IT"], "\n")

# 5. Employees with Salary > 5000
print("Employees with Salary > 5000:\n", df[df["Salary"] > 5000], "\n")

# 6. Employees with 5–10 years of Experience
print("Employees with 5–10 years experience:\n", df[df["Experience"].between(5, 10)], "\n")

# 7. Employees based in New York or Chicago
print("Employees in New York or Chicago:\n", df[df["City"].isin(["New York", "Chicago"])], "\n")

# 8. Retrieve employee details using .loc[] for Employee_ID = 104
print("Employee with ID 104:\n", df.loc[df["Employee_ID"] == 104], "\n")

# 9. Select 2nd to 5th employees using .iloc[]
print("2nd to 5th employees:\n", df.iloc[1:5], "\n")

# 10. Change index to Employee_ID
df_indexed = df.set_index("Employee_ID")
print("DataFrame with Employee_ID as index:\n", df_indexed, "\n")

# 11. Reset index back to default
df_reset = df_indexed.reset_index()
print("DataFrame with reset index:\n", df_reset, "\n")

# 12. Employees earning > 7000 in Finance Department
print("Finance employees with Salary > 7000:\n", df[(df["Salary"] > 7000) & (df["Department"] == "Finance")], "\n")

# 13. Employees Age > 30 and Experience < 5 years
print("Employees with Age > 30 and Experience < 5:\n", df[(df["Age"] > 30) & (df["Experience"] < 5)], "\n")
```
