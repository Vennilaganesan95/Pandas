```python
import pandas as pd

# Sample Employee Salary Data
data = {
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen"],
    "Department": ["HR", "Finance", "IT", "Finance", "IT", "HR", "IT", "Finance"],
    "Salary": [40000, 55000, 60000, 45000, 75000, 38000, 72000, 50000]
}

df = pd.DataFrame(data)
print(df)

# Categorize employees into Low / Medium / High salary groups
df["Salary_Category"] = df["Salary"].apply(
    lambda x: "Low" if x < 45000 else ("Medium" if x < 65000 else "High")
)

print(df[["Employee", "Salary", "Salary_Category"]])

dept_summary = df.groupby("Department").agg(
    Average_Salary=("Salary", "mean"),
    Max_Salary=("Salary", "max"),
    Min_Salary=("Salary", "min"),
    Employee_Count=("Employee", "count")
).reset_index()

print(dept_summary)

pivot_table = pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    columns="Salary_Category",
    aggfunc="count",
    fill_value=0
)

print(pivot_table)
sorted_employees = df.sort_values(by="Salary", ascending=False)
print(sorted_employees)



```
